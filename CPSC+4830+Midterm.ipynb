```python
Outlook=['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast','Sunny','Sunny','Rain','Sunny','Overcast','Overcast','Rain']
Temp=['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']
Humidity=['High','High','High','High','Normal','Normal','Normal','High','Normal','Normal','Normal','High','Normal','High']
Wind=['Weak','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Strong']
Play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
```

```python
#df=pd.DataFrame(Outlook,Temp,Humidity,Wind,Play)
df['outlook']=Outlook
df['temp']=Temp
df['humidity']=Humidity
df['wind']=Wind
df['Play']=Play
df
```

```python
#function to return gini_index
def decisionTree(df, rootClass):
    count_row = df.shape[0]
    k=[]
    z=[]
    gini_index=0
    for i in range(df[rootClass].nunique()):
        k.append(df[rootClass].value_counts()[i])
        z.append(round(k[i]/count_row,3))
        gini_index = gini_index + (z[i])**2
    return round(1-gini_index,3)
```

```python
Outlook_gini=decisionTree(df,'Play')
Outlook_gini
```

```python
outlook_Sunny=df[df['outlook']=='Sunny']
gini_index_outlook_Sunny=decisionTree(outlook_Sunny,'Play')
gini_index_outlook_Sunny
```

```python
outlook_Overcast=df[df['outlook']=='Overcast']
gini_index_outlook_Overcast=decisionTree(outlook_Overcast,'Play')
gini_index_outlook_Overcast
```

```python
outlook_Rain=df[df['outlook']=='Rain']
gini_index_outlook_Rain=decisionTree(outlook_Rain,'Play')
gini_index_outlook_Rain
```

```python
Outlook_overall_giniIndex= Outlook_gini-(((len(df[df['outlook']=='Sunny'])/len(df))*gini_index_outlook_Sunny + 
                                        ((len(df[df['outlook']=='Overcast'])/len(df))*gini_index_outlook_Overcast +
                                         ((len(df[df['outlook']=='Rain'])/len(df))*gini_index_outlook_Rain)
                                       
```

### Q2. The SMS Spam Collection is a set of SMS tagged messages that have been collected from SMS Spam research database. It contains one set of SMS messages in English of 5,572 messages, tagged acording being ham (legitimate) or spam.

### Can you use this dataset to build a prediction model that will accurately classify which texts are spam? Use NLTK with your choice of classifier. [10 Marks]

### Out of the 5572 sets of sms messages, approximately 20% of the messages have been trimmed and kept aside for Evaluating your model. 

```python
import string
string.punctuation
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
spamham = pd.read_csv("spamhamtrain.csv")
spamham.head()
```

```python
#function to remove punctiation and stop words
def message_cleaning(message):
    Test_punc_removed = [w for w in message if w not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join_clean = [word for word in Test_punc_removed_join.split() 
                                    if word.lower() not in stopwords.words('english')]
    return Test_punc_removed_join_clean
```

```python
#spamham_clean = spamham['Message'].apply(message_cleaning)
```

```python
# creating count vectorixer
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer = message_cleaning)
spamham_countvectorizer = vectorizer.fit_transform(spamham['Message'])
```

```python
print(vectorizer.get_feature_names())
```

```python
print(spamham_countvectorizer.toarray())  
```

```python
# divide data in testing and training
X = spamham_countvectorizer
y = spamham['Category']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Using Naive Byes Classifier

```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

# Evaluating Model

```python
from sklearn.metrics import classification_report, confusion_matrix
y_predict_test = NB_classifier.predict(X_test)
print(classification_report(y_test, y_predict_test))
```

```python
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True)
```
