```python
import numpy as np
import pandas as pd
import math
```

```python
df = pd.read_csv('HandCalculation.csv')
```

```python
df
```

```python
df.columns = df.columns.str.replace(' ','').str.lower()
df
```

```python
def decisionTree(df, rootClass):
    count_row = df.shape[0]
    k=[]
    z=[]
    finalCal=[]
    entropy=0
    gini_index=0
    for i in range(df[rootClass].nunique()):
        k.append(df[rootClass].value_counts()[i])
        z.append(round(k[i]/count_row,3))
        entropy = entropy - (z[i] * math.log(z[i],2))
        gini_index = gini_index + (z[i])**2
    clstn_err = round(1 - max(z),3)
    finalCal.append(round(entropy,3))
    finalCal.append(round(1-gini_index,3))
    finalCal.append(clstn_err)
    return finalCal
```

```python
overall=decisionTree(df,'transportationmode')
overall
```

```python
df_genderMale=df[df['gender']=='Male']
overall=decisionTree(df_genderMale,'transportationmode')
overall
```
